```python
import random
import copy

#Create a "current" matrix of living and dead cells
rows, cols = (38, 38)
current = [[0] * cols for _ in range(rows)]

#First block
current[5][1]=1
current[6][1]=1
current[5][2]=1
current[6][2]=1

#First circle
current[3][13]=1
current[3][14]=1

current[4][12]=1
current[4][16]=1

current[5][11]=1
current[5][17]=1

current[6][11]=1
current[6][15]=1
current[6][17]=1
current[6][18]=1

current[7][11]=1
current[7][17]=1

current[8][12]=1
current[8][16]=1

current[9][13]=1
current[9][14]=1

#Second circle
current[1][25]=1

current[2][23]=1
current[2][25]=1

current[3][21]=1
current[3][22]=1

current[4][21]=1
current[4][22]=1

current[5][21]=1
current[5][22]=1

current[6][23]=1
current[6][25]=1

current[7][25]=1

#rightmost block
current[3][35]=1
current[3][36]=1
current[4][35]=1
current[4][36]=1

#frames for animation
frames=[]

frames.append(copy.deepcopy(current))

#Create a copy called "updated"
updated = [[0] * cols for _ in range(rows)]

#First block
updated[5][1]=1
updated[6][1]=1
updated[5][2]=1
updated[6][2]=1

#First circle
updated[3][13]=1
updated[3][14]=1

updated[4][12]=1
updated[4][16]=1

updated[5][11]=1
updated[5][17]=1

updated[6][11]=1
updated[6][15]=1
updated[6][17]=1
updated[6][18]=1

updated[7][11]=1
updated[7][17]=1

updated[8][12]=1
updated[8][16]=1

updated[9][13]=1
updated[9][14]=1

#Second circle
updated[1][25]=1

updated[2][23]=1
updated[2][25]=1

updated[3][21]=1
updated[3][22]=1

updated[4][21]=1
updated[4][22]=1

updated[5][21]=1
updated[5][22]=1

updated[6][23]=1
updated[6][25]=1

updated[7][25]=1

#rightmost block
updated[3][35]=1
updated[3][36]=1
updated[4][35]=1
updated[4][36]=1

Tsteps=100

#evolve the system through each timestep
for t in range(Tsteps):

    #Loop through current and update the matrix "updated" accordingly
    for i in range(rows):
        for j in range(cols):

            sum=0

            #Look at the 6 nearest neighbors

            #above
            if i>0 and current[i-1][j]==1:
                sum=sum+1
            #left
            if j>0 and current[i][j-1]==1:
                sum=sum+1

            #down
            if i<rows-1 and current[i+1][j]==1:
                sum=sum+1
            #right
            if j<cols-1 and current[i][j+1]==1:
                sum=sum+1

            #upper left
            if i>0 and j>0 and current[i-1][j-1]==1:
                sum=sum+1
            #lower right
            if i<rows-1 and j<cols-1 and current[i+1][j+1]==1:
                sum=sum+1

            #bottom left
            if i<rows-1 and j>0 and current[i+1][j-1]==1:
                sum=sum+1
            #upper right
            if i>0 and j<cols-1 and current[i-1][j+1]==1:
                sum=sum+1

            #any live cell with <2 neighbors dies by underpopulation
            if sum<2 and current[i][j]==1:
                updated[i][j]=0
            
            #any live cell with 2 or 3 live neighbors survive
            elif current[i][j]==1 and (sum==2 or sum==3):
                updated[i][j]=1
                
            #any live cell with >3 neighbors dies by overpopulation
            elif current[i][j]==1 and sum>3:
                updated[i][j]=0
                
            #any dead cell with exactly 3 neighbors becomes alive by reproduction
            elif current[i][j]==0 and sum==3:
                updated[i][j]=1
                
    for i in range(rows):
        for j in range(cols):
            
            current[i][j]=updated[i][j]
    
    # create a deep copy of updated
    updated_copy = copy.deepcopy(updated)
    
    # append the copy to frames
    frames.append(updated_copy)
```

```python
import matplotlib.pyplot as plt

#Animation cell 
%matplotlib

plt.clf()

for i in range(len(frames)):

        for x in range(rows):
            for y in range(cols):

                if frames[i][x][y]==1:

                    plt.plot(y,rows-x,'g*')
                    ax = plt.gca()
                    ax.set_facecolor('k')

                else:

                    plt.plot(y,rows-x,'k*')
                    ax = plt.gca()
                    ax.set_facecolor('k')

        plt.show()
        plt.pause(0.1)

        if i<len(frames)-1:
            plt.clf()
```
